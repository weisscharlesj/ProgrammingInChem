# Part A
Say we have a two-step reaction:

$ A \rightarrow B \rightarrow C $

that is first order for all reactants/products, and we have an initial concentration of $[A]_0 = 3.6$~M, and no $B$ or $C$ present initially.  Additionally, assume we have rate constants for the first and second steps as 1.0 1/s.

As described in the lab handout, we can calculate the concentrations of $A, B$, and $C$ as

$ [A]_i = [A]_{(i-1)} + \frac{\Delta [A]}{\Delta t}\times \Delta t = [A]_{(i-1)} - k_1 [A]_{(i-1)} \Delta t$

And similarily for be (accounting for the growth of $B$ from decay of $A$, and then the decay of $B$ to $C$):

$ [B]_i = [B]_{(i-1)} + k_1 [A]_{(i-1)} \Delta t - k_2[B]_{(i-1)} \Delta t$

And for $C$:

$[C]_i = [C]_{(i-1)} + k_2[B]_{(i-1)} \Delta t$

Below, the numerical solution for this process is given.  Note that $timesteps$ is an array of times that you can/should change.  The array has a start, stop, and step.  So, initially, $timesteps = np.arange(0,10,0.1)$ would have a start time of 0 seconds, stop time of 10 seconds, and 0.1 second steps. 


```python
#libraries to import 
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit


A0 = 3.6 # M
B0 = 0.0 # M
C0 = 0.0 # M
k1 = 1.0 # 1/sec
k2 = 1.0 # 1/sec

timesteps = np.arange(0,10,0.1)

A = np.zeros(np.size(timesteps))
B = np.zeros(np.size(timesteps))
C = np.zeros(np.size(timesteps))
A[0] = A0
B[0] = B0
C[0] = C0
A_order = 1
B_order = 1

for i in range(np.size(timesteps)):
    if i > 0:
        A[i] = A[i-1] - k1*A[i-1]**(A_order)*(timesteps[i]-timesteps[i-1])
        B[i] = B[i-1] + k1*A[i-1]**(A_order)*(timesteps[i]-timesteps[i-1]) - k2*B[i-1]**(B_order)*(timesteps[i]-timesteps[i-1])
        C[i] = C[i-1] + k2*B[i-1]**(B_order)*(timesteps[i]-timesteps[i-1])
        
plt.plot(timesteps,A,color='green')
plt.plot(timesteps,B,color='red')
plt.plot(timesteps,C,color='blue')

plt.ylabel("Concentration (M)")
plt.xlabel("time (s)")
plt.title("Title")
```

$ A \rightarrow 2B \rightarrow C $

Notice, you have to account for the second order reaction order for $B$, and also the fact that 2 $B$ molecules are used in each reaction.  Make sure to format your plot well.

```python


import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Say we have a two-step reaction:
# A -> B -> C
# that is first order for A->B, but second order for B->C order for all reactants/products, and we have
A0 = 3.6 # M
B0 = 0.0 # M
C0 = 0.0 # M
k1 = 1.0 # 1/sec
k2 = 1.0 # 1/sec

# now,
timesteps = np.arange(0,10,0.01)
A = np.zeros(np.size(timesteps))
B = np.zeros(np.size(timesteps))
C = np.zeros(np.size(timesteps))
A[0] = A0
B[0] = B0
C[0] = C0
A_order = 1
B_order = 2

for i in range(np.size(timesteps)):
    if i > 0:
        A[i] = A[i-1] - k1*A[i-1]**(A_order)*(timesteps[i]-timesteps[i-1])
        B[i] = B[i-1] + k1*A[i-1]**(A_order)*(timesteps[i]-timesteps[i-1])*2- k2*B[i-1]**(B_order)*(timesteps[i]-timesteps[i-1])*2
        C[i] = C[i-1] + k2*B[i-1]**(B_order)*(timesteps[i]-timesteps[i-1])
        
plt.plot(timesteps,A,color='orange')
plt.plot(timesteps,B,color='cyan')
plt.plot(timesteps,C,color='purple')
```
